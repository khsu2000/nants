# Data Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

## Single Categorical Variable: Bar Chart

```python
bar_chart = sns.countplot(
    x = "who",
    data = titanic
)
plt.title("Distribution of Demographics")
plt.xlabel("category")
plt.show()
```

## Single Numerical Variable: Histogram

```python
sns.distplot(titanic.age)
plt.title("Distribution of Age")
plt.ylabel("density")
plt.show()
```

## Multiple Categorical Variables: Multi-Level Bar Chart

```python
sns.countplot(
    x = "who",
    hue = "class",
    data = titanic
)
plt.title("Distribution of Demographics")
plt.xlabel("category")
plt.show()
```

## Two Numerical Variables: Scatter Plot

```python
sns.scatterplot(
    x = "sepal_length",
    y = "sepal_width",
    data = iris,
)
plt.title("Sepal Width vs. Sepal Length")
plt.show()
```

## Two Numerical Variables: Hex Bins

```python
sns.scatterplot(
    x = "horsepower",
    y = "mpg",
    data = mpg
)
plt.title("Acceleration vs. Horsepower")
plt.show()
```

```python
sns.jointplot(
    data = mpg, 
    x = "horsepower", 
    y = "mpg",
    kind = "hex"
)
plt.show()
```

## Two Numerical Variables: Line Plots

```python
sns.lineplot(
    x = "timepoint", 
    y = "signal",
    data = fmri
)
plt.title("Signal over Time")
plt.show()
```
