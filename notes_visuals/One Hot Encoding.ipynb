# One Hot Encoding

```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

## Encoding Class Column

```python
titanic_class = pd.DataFrame(titanic["class"])
titanic_class
```

```python
encoded_class = pd.get_dummies(titanic_class)
encoded_class.head()
```

```python
titanic.drop("class", axis = 1, inplace = True)
titanic = pd.concat([titanic, encoded_class], axis = 1)
titanic.head()
```
